## Using regex with the "contains" method in Pandas

In addition to just matching on a regular substring, we can also use `contains` to match on regular extressions. We will use the exact format as before, except this time let's use a bit of regex to only find tchaou

```python
import re

regexp = re.compile(r"sen")


matches = []
for game in df['Titre']:
    if re.search(regexp, game):
        matches.append(game)

print(matches)
```
